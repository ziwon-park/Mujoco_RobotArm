```python
def read_and_visualize(folder_path):
    for i in range(1, 8):  
        file_path = os.path.join(folder_path, f"free_joint_{i}.csv")
        data = pd.read_csv(file_path, sep=',')
        plt.figure(figsize=(10, 6))
        plt.plot(data.iloc[:, 0], data.iloc[:, 1], label=f'Joint {i}')
        plt.xlabel('Sequence Number')
        plt.ylabel('Q value')
        plt.title(f'Joint {i} Q Value Over Time')
        plt.legend()
        plt.show()
```

```python
def read_and_visualize_single_file(folder_path, n):
    file_path = os.path.join(folder_path, f"free_joint_1.csv")
    data = pd.read_csv(file_path, header=None)

    nth_data = data.iloc[n-1]

    plt.figure(figsize=(10, 6))
    plt.plot(nth_data)
    plt.xlabel('Time Step')
    plt.ylabel('Q value')
    plt.title(f'Q Value Changes Over Time for Row {n}')
    plt.legend()
    plt.show()
```

```python
folder_path = "/home/robros/model_uncertainty/script/dataset_generator/csv/temp/temp"
n=1
read_and_visualize_single_file(folder_path, n)
```
